- 从行业来看：

- 第三梯队：清洁用品-烟草类，清洁用品、邮电通讯、交通行业以微小差距领先，但与本梯队其他行业间的差距不大

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType


def geo_卫视品牌赞助数量() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("各卫视品牌赞助数量统计",卫视品牌赞助数量,type_=ChartType.HEATMAP,)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(max_=31),
            title_opts=opts.TitleOpts(title="2019年下半年（6-11月）各地方卫视品牌赞助商数量总计"),
        )
    )
    return c
图2=geo_卫视品牌赞助数量()
图2.render_notebook()
```
