# 1. 2019年下半年各行业赞助情况总计（6-11月赞助总和）

```python
import pandas as pd
df1 = pd.read_csv('brandcount.csv')
df1.head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Pie,Grid

industry = list(df1['行业类别'])
data = list(df1['6-11月'])

def pie_scroll_legend() -> Pie:
    c = (
        Pie(init_opts=opts.InitOpts(width="600px",height="400px"))
        .add("",[list(z) for z in zip(industry,data)])
        .set_global_opts(
            title_opts=opts.TitleOpts(title="2019年下半年各行业赞助（地方卫视）时长总计",pos_left="center"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    grid=Grid()
    # 仅使用pos_top修改相对顶部的位置
    grid.add(c,grid_opts=opts.GridOpts(pos_top="50%"))

    return grid

a=pie_scroll_legend()
#a.render('brand.html')
a.render_notebook()
```

- 饮料业赞助节目时长最长，药品、食品热忱相对较高

### 从月份来看：饮料与食品
- 6-8月饮料赞助呈上升趋势，夏季饮料市场较为活跃，刚需较大；而食品类在经历7月份大跌后，到11月间一路稳步上升
- 9-11月食品赞助攀升，秋冬季人类对食物愈发没有抵抗力，品牌商抓住这一季节痛点精准营销，实力诱使受众养嫖过冬

### 从行业来看：行业赞助情况呈“三段式”分布
- 根据图表各行业数值差异，各赞助行业可以划分为三大梯队。
- 第一梯队：饮料、药品、食品荣获赞助行业第一梯队，从2019年下半年总览来看，这三个行业的赞助时长远高于其余行业，差距明显
- 第二梯队：酒类到衣着这一部分行业为第二梯队，从6-11月下半年总汇数据看，第二梯队行业的差异呈现一定的等级化，且总体与第一梯队形成一倍的差距
- 第三梯队：清洁用品到烟草类为行业第三梯队，清洁用品、邮电通讯、交通行业以微小差距领先，但与本梯队其他行业间的差距不大



# 3. 2019下半年31个地方卫视品牌赞助商数量统计

```python
df2 = pd.read_csv('channel.csv')
df2.head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType


def geo_卫视品牌赞助数量() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("各卫视品牌赞助数量统计",list(zip(list(df2['省份']),list(df2['品牌数量']))),type_=ChartType.HEATMAP,)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(max_=31),
        )
    )
    return c
图2=geo_卫视品牌赞助数量()
#图2.render('brandcount.html')
图2.render_notebook()
```

- 依据热力图，东部城市及沿海城市品牌赞助商数量普遍多于内陆及西部地区
- 深圳、江苏、浙江、北京、广东五个地方卫视的赞助商数量名列前茅，其中江苏卫视、浙江卫视为一线卫视，北京卫视为二线卫视，深圳卫视、广东卫视为三线卫视（以各卫视历年收视率为划分依据）
- 注：赞助商数量热力图只作为相关参考图例之一（只是满足地图交互），赞助商数量的多少可以反映一个卫视的发展活跃情况，但并不可以完全反映，还有其他影响指标如赞助商的质量、知名度、商业价值等。
